```python
def display_all_surveys():
    ref = db.reference('surveys')
    all_surveys = ref.get()
    if all_surveys:
        print("\n--- All Survey Records ---")
        for survey_id, survey_data in all_surveys.items():
            print(f"ID: {survey_id}, Data: {survey_data}")
        print("------------------------")
    else:
        print("No surveys found in the database.")
```

```python
def add_survey():
    ref = db.reference('surveys')

    current_surveys = ref.get()
    if current_surveys:
        max_id = max([int(k) for k in current_surveys.keys()])
        new_id = max_id + 1
    else:
        new_id = 1

    print(f"\n--- Adding New Survey (ID: {new_id}) ---")
    age = int(input("Enter age: "))
    gender = input("Enter gender: ")
    grade_level = input("Enter grade level: ")
    section = input("Enter section: ")
    q1_satisfaction = int(input("Enter Q1 satisfaction (1-5): "))
    q2_usability = int(input("Enter Q2 usability (1-5): "))
    q3_design = int(input("Enter Q3 design (1-5): "))
    q4_clarity = int(input("Enter Q4 clarity (1-5): "))
    q5_recommend = int(input("Enter Q5 recommend (1-5): "))
    q6_overall = int(input("Enter Q6 overall (1-5): "))
    q7_comments = input("Enter Q7 comments: ")

    new_survey_data = {
        "respondent_id": new_id,
        "age": age,
        "gender": gender,
        "grade_level": grade_level,
        "section": section,
        "q1_satisfaction": q1_satisfaction,
        "q2_usability": q2_usability,
        "q3_design": q3_design,
        "q4_clarity": q4_clarity,
        "q5_recommend": q5_recommend,
        "q6_overall": q6_overall,
        "q7_comments": q7_comments
    }

    ref.child(str(new_id)).set(new_survey_data)
    print(f"Survey with ID {new_id} added successfully!")
```

```python
def display_all_surveys():
    ref = db.reference('surveys')
    all_surveys = ref.get()
    if all_surveys:
        print("\n--- All Survey Records ---")
        # Handle both dictionary and list types for all_surveys
        if isinstance(all_surveys, dict):
            for survey_id, survey_data in all_surveys.items():
                print(f"ID: {survey_id}, Data: {survey_data}")
        elif isinstance(all_surveys, list):
            for survey_data in all_surveys:
                if survey_data:  # Add check for NoneType elements in the list
                    # Assuming each item in the list is a dictionary with a 'respondent_id'
                    survey_id = survey_data.get('respondent_id', 'N/A')
                    print(f"ID: {survey_id}, Data: {survey_data}")
                else:
                    print("Found a None entry in the surveys list.")
        else:
            print(f"Unexpected data format for surveys: {type(all_surveys)}")
        print("------------------------")
    else:
        print("No surveys found in the database.")

def main_menu():
    while True:
        print("\n--- Firebase Survey Database Menu ---")
        print("1. Display All Surveys")
        print("2. Add New Survey")
        print("3. Update Survey (Not Implemented)")
        print("4. Delete Survey (Not Implemented)")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        match choice:
            case '1':
                display_all_surveys()
            case '2':
                add_survey()
            case '3':
                print("Update function is not yet implemented.")
            case '4':
                print("Delete function is not yet implemented.")
            case '5':
                print("Exiting program. Goodbye!")
                break
            case _:
                print("Invalid choice. Please enter a number between 1 and 5.")

main_menu()
```
